```python
# Data
company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

employee_detail_list = [
    {
        'name': 'EMP-0001',
        'first_name': 'John',
        'last_name': 'Doe',
        'full_name': 'John Doe',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0002',
        'first_name': 'Tom',
        'last_name': 'Smith',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0003',
        'first_name': 'Andrew',
        'last_name': 'Sebastian',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0005',
        'first_name': 'Ying Han',
        'last_name': 'Tan',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0015',
        'first_name': 'Kenneth',
        'last_name': 'Ng',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0018',
        'first_name': 'Rubby',
        'last_name': 'Lee',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
        'nationality': 'Hong Kong'
    },
    {
        'name': 'EMP-0017',
        'first_name': 'Robert',
        'last_name': 'White',
        'full_name': 'Robert White',
        'company': 'Company 1',
        'nationality': 'United Kingdom'
    }
]

# Task 1
def get_sorted_companies():
    sorted_companies = sorted(company_detail_list, key=lambda x: x['name'], reverse=True)
    return [{'name': company['name']} for company in sorted_companies]

# Task 2
def print_company_domains():
    for company in company_detail_list:
        print(f"{company['name']}: {company['domain']} ({company['country']})")

# Task 3
def list_employees_by_domain():
    domain_employees = {}
    for company in company_detail_list:
        domain = company['domain']
        domain_employees[domain] = []

    for employee in employee_detail_list:
        company_name = employee['company']
        for company in company_detail_list:
            if company['name'] == company_name:
                domain_employees[company['domain']].append(employee['full_name'])
                break

    return domain_employees

# Task 4
def get_employees_with_country():
    employees_with_country = []
    for employee in employee_detail_list:
        company_name = employee['company']
        for company in company_detail_list:
            if company['name'] == company_name:
                employees_with_country.append({
                    'full_name': employee['full_name'],
                    'company': company_name,
                    'country': company['country']
                })
                break
    return employees_with_country

# Task 5
def get_employee_nationality_count():
    nationality_count = []
    for company in company_detail_list:
        company_name = company['name']
        employee_nationality = {}
        for employee in employee_detail_list:
            if employee['company'] == company_name:
                nationality = employee['nationality']
                if nationality in employee_nationality:
                    employee_nationality[nationality] += 1
                else:
                    employee_nationality[nationality] = 1
        nationality_count.append({
            'company': company_name,
            'employee_nationality': employee_nationality
        })
    return nationality_count

# Running the tasks
print("Task 1 - Sorted Companies:")
print(get_sorted_companies())
print("\nTask 2 - Company Domains:")
print_company_domains()
print("\nTask 3 - Employees by Domain:")
print(list_employees_by_domain())
print("\nTask 4 - Employees with Country:")
print(get_employees_with_country())
print("\nTask 5 - Employee Nationality Count:")
print (get_employee_nationality_count())
```
